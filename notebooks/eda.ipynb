```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\A\OneDrive\Desktop\developer arena task\task8\data\cleaned_data.csv")
```

```python
df.columns
```

```python
sns.histplot(df["Total_Sales"], kde=True)

plt.title("Sales Distribution")
plt.show()
```

```python
region_sales = df.groupby("Region")["Total_Sales"].sum()

region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.show()
```

```python
sns.scatterplot(x=df['Product'], y=df["Total_Sales"])

plt.title("Marketing vs Sales")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)

plt.title("Correlation Matrix")
plt.show()
```
