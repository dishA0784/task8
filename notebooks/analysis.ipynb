```python
import pandas as pd
from scipy import stats
from sklearn.linear_model import LinearRegression


df = pd.read_csv(r"C:\Users\A\OneDrive\Desktop\developer arena task\task8\data\cleaned_data.csv")



corr, p = stats.pearsonr(df["Quantity"], df["Total_Sales"])

print("Correlation:", corr)
print("p-value:", p)



if p < 0.05:
    print("Marketing significantly affects sales")
else:
    print("No significant effect")



X = df[["Quantity"]]
y = df["Total_Sales"]

model = LinearRegression()

model.fit(X, y)

print("Coefficient:", model.coef_[0])
print("Intercept:", model.intercept_)
```
